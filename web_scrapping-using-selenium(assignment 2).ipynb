```python
# importing required libraries
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
# initializing driver
driver=webdriver.Chrome(r"C:\Users\grgro\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
# function to get flipkart webpage
def get_flipkart():
    driver.get("https://www.flipkart.com/")
get_flipkart()
```

```python
# function to remove popup ads
def remove_popup():
    try:
        pop_up=driver.find_element(By.XPATH,'/html/body/div[2]/div/div/button')
        pop_up.click()
    except:
        pass
remove_popup()
```

```python
# giving value in search box
designation=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
designation.send_keys('sunglasses')
```

```python
# function to click search button
search=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
search.click()
```

```python
# Initializing empty list to store the values
brand_names=[]
product_description=[]
product_price=[]
```

```python
# initializing start and end to get data from three pages
start=0
end=3
for page in range(start,end):
#     finding all the brand name and appending the value in the list
    brand=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in brand:
        brand_name=i.text
        brand_names.append(brand_name)
        if len(brand_names)<100:
            continue
        else:
            break
        
#         finding all the description and appending the value in the list
    product_desc=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for desc in product_desc:
        description=desc.text
        product_description.append(description)
        if len(product_description)<100:
            continue
        else:
            break 
        
#         finding all the description and appending the value in the list
    product_prices=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for price in product_prices:
        prices=price.text
        product_price.append(prices)
        if len(product_price)<100:
            continue
        else:
            break 
        
#         function to click next button
    next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(3)
# len(brand_names)
# len(product_description)
# len(product_price)
```

```python
# checking the length of the list
print(len(brand_names),len(product_description),len(product_price))
```

```python
# creating the dataframe 
sunglass_df=pd.DataFrame({'Brand names':brand_names,'Product description':product_description,'Prices':product_price})
sunglass_df
```

```python
# initializing the empty list to store the values
shoe_brand_names=[]
shoe_description=[]
shoe_price=[]
```

```python
# functions to scrap sneakers data
def sneakers_data():
    start=0
    end=3
    for page in range(start,end):
        shoe_names=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
        for shoe in shoe_names:
            shoe_brand_names.append(shoe.text)
            if len(shoe_brand_names)<100:
                continue
            else:
                break

        shoe_dec=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
        for description in shoe_dec:
            shoe_description.append(description.text)
            if len(shoe_description)<100:
                continue
            else:
                break 


        shoes_prices=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
        for price in shoes_prices:
            shoe_price.append(price.text)
            if len(shoe_price)<100:
                continue
            else:
                break 

        next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
        next_button.click()
        time.sleep(3)
sneakers_data()
```

# Data scraping for top1000 quotes

```python
driver.get('https://www.azquotes.com/')
```

```python
top_quotes=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[1]/div/div[3]/ul/li[5]/a')
top_quotes.click()
```

```python
quotes_title=[]
author_names=[]
types_of_quotes=[]
```

```python
def scrap_quotes():
    start=0
    end=10
    for page in range(start,end):
        titles=driver.find_elements(By.XPATH,'//a[@class="title"]')
        for title in titles:
            quotes_title.append(title.text)
        
        authors=driver.find_elements(By.XPATH,'//div[@class="author"]')
        for author in authors:
            author_names.append(author.text)
            
        tags=driver.find_elements(By.XPATH,'//div[@class="tags"]')
        for tag in tags:
            types_of_quotes.append(tag.text)
            
        
        next_button=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div/div/div/div[1]/div/div[3]/li[12]')
        next_button.click()
        time.sleep(3)
scrap_quotes()
```

```python
print(len(quotes_title),len(author_names),len(types_of_quotes))
```

```python
quotes_df=pd.DataFrame({'Quote':quotes_title,'Author name':author_names,'Tags':types_of_quotes})
quotes_df
```
