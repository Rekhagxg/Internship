# scrapping Multiple  Title

```python
price=[] #empty list for store 

for i in soup.find_all('span',class_="cfpn1d"):
    price.append(i.text)
```

```python
price
```

```python
name=[] #empty list for store 

for i in soup.find_all('h3',):
    name.append(i.text)
```

```python
name
```

```python
images=[]
for i in soup.find_all('img',class_=''):
    images.append(i.get('src'))
images
```
