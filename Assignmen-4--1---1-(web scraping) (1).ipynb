# Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/
You have to find the following details:
A) Dataset name B) Data type
C) Task
D) Attribute type E) No of instances F) No of attribute G) Year
Note: - from the home page you have to go to the ShowAllDataset page through code.


```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.action_chains import ActionChains
```

```python
#connect to driver
driver=webdriver.Chrome(r"/Users/rekhagrg/Downloads/chromedriver")
```

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
try:
    datasets=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/header/nav/ul/li[1]')
    datasets.click()
except:
    print('Error!!')


```

```python
try:
    expand=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]')
    expand.click()
except:
    print("Error")
```

```python
dataset_name=[]
data_type=[]
task=[]
attribute_type=[]
no_of_instances=[]
no_of_attribute=[]
year=[]
```

```python
start=0
end=20

count=1
for page in range(start,end):
    names=driver.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]')
    for name in names:
        dataset_name.append(name.text)
    
    for d_type in range(1,count+2):
        types=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div["+str(count)+"]/div[2]/div/table/tbody/tr/td[2]')
        tasks=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div["+str(count)+"]/div[1]/div[2]/div/div[1]/span')
        d_type=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div["+str(count)+"]/div[1]/div[2]/div/div[2]/span')
        instances=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div["+str(count)+"]/div[1]/div[2]/div/div[3]/span')
        attributes=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div["+str(count)+"]/div[1]/div[2]/div/div[4]/span')
        dates=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div["+str(count)+"]/div[2]/div/table/tbody/tr/td[3]')

        
    for t in types:
        attribute_type.append(t.text)
        
    for tas in tasks:
        task.append(tas.text)
        
    for d in d_type:
        data_type.append(d.text)

    for instance in instances:
        no_of_instances.append(instance.text)

    for attribute in attributes:
        no_of_attribute.append(attribute.text)

    for date in dates:
        year.append(date.text)
            
    next_button = driver.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
    actions = ActionChains(driver)
    actions.move_to_element(next_button).click().perform()
    time.sleep(2)
```

```python
print(len(dataset_name),len(attribute_type),len(task),len(data_type),len(no_of_instances),len(no_of_attribute),len(year))
```

```python
uci_df=pd.DataFrame({'Dataset name':dataset_name,'Data type':data_type,'Task':task,'Attribute type':attribute_type,'No of instances':no_of_instances,'No of attribute':no_of_attribute,'Date donated':year})
uci_df
```
